```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe = pd.read_csv('/content/Pokemon.csv')
print(dataframe)
```

```python
#Strongest 5 pokemons

strongestpokes = dataframe.sort_values(by=["Total"]).iloc[::-1,:].head()
print(strongestpokes)
```

```python
#Weakest 5 pokemons

weakestpoke = dataframe.sort_values(by=["Total"]).head()
print(weakestpoke)
```

```python
#Strongest and weakest pokemons with Scatter Plot.

weakestpoke = dataframe.sort_values(by=["Total"]).head()

strongestpokes = dataframe.sort_values(by=["Total"]).iloc[::-1,:].head()

newpoke = pd.concat([weakestpoke, strongestpokes] ,axis=0)

plt.scatter(x=newpoke.Name,y=newpoke.Total)
plt.xticks(rotation=75)
plt.yticks(np.arange(0,1000,40))
plt.show()
```

```python
#To get correlation between every two float variables

numericdataframe = dataframe.iloc[:,4:11]
print(numericdataframe)

print("Correlation between every two colums")
numericdataframe.corr()
```

```python
# correlation between variables
sns.heatmap(numericdataframe.corr(),linewidths=2,annot = True)
```

```python
#Line plot between sp attack and total

plt.subplot(2,1,1)
plt.plot(dataframe["Sp. Atk"],label="Special Attack", color = "orange")
plt.xlabel("Indexes")
plt.ylabel("Special Attack")
plt.legend()

plt.subplot(2,1,2)
plt.plot(dataframe.Total,label="Total")
plt.xlabel("Indexes")
plt.ylabel("Total")
plt.legend()

plt.tight_layout()
plt.show()
```
